```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Parse the data
data = {
    "acyl halides": 0.7805,
    "alcohols": 0.9466,
    "aldehydes": 0.9425,
    "alkane": 0.9602,
    "alkene": 0.8529,
    "alkyl halides": 0.8286,
    "alkyne": 0.9425,
    "amides": 0.6875,
    "amines": 0.8735,
    "aromatics": 0.9807,
    "carboxylic acids": 0.9077,
    "esters": 0.9042,
    "ether": 0.9193,
    "ketones": 0.8659,
    "methyl": 0.9615,
    "nitriles": 0.6752,
    "nitro": 0.967,
    "weighted_f1": 0.931
}

# Extract weighted_f1 and create DataFrame excluding it
weighted_f1 = data.pop("weighted_f1")
df = pd.DataFrame(list(data.items()), columns=['Functional Group', 'F1 Score'])
df = df.sort_values('F1 Score', ascending=False)

# Set up the figure with a nice size
plt.figure(figsize=(12, 8))

# Create the horizontal bar plot
bars = plt.barh(df['Functional Group'], df['F1 Score'], color='#4285F4')

# Add value labels at the end of each bar
for i, bar in enumerate(bars):
    plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2, 
             f'{df["F1 Score"].iloc[i]:.3f}', 
             va='center', fontsize=10)

# Add a vertical line for weighted average
plt.axvline(weighted_f1, color='#DB4437', linestyle='--', linewidth=2, 
            label=f'Weighted F1: {weighted_f1:.3f}')

# Set labels and title
plt.xlabel('F1 Score', fontsize=12)
plt.title('F1 Scores by Functional Group (NIST)', fontsize=14, fontweight='bold')

# Add grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Set x-axis limits
plt.xlim(0, 1.05)

# Add a legend for the weighted average line
plt.legend(loc='lower right')

# Tight layout
plt.tight_layout()

# Show the plot
plt.show()

```
